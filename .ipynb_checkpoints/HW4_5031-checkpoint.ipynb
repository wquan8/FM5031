$f_L(x)=\frac{\theta_L}{\beta_L}(1 - \xi \frac{x-\eta_L}{\beta_L})^{-1-\frac{1}{\xi}})$

$f_U(x)=\frac{\theta_U}{\beta_U}(1 + \xi \frac{x-\eta_U}{\beta_U})^{-1-\frac{1}{\xi}})$

(1) $K \le \eta_L$

$\frac{p(T,K)}{d(T)} =\int_{\eta_L+\frac{\beta_L}{\xi_L}}^{K} (K-s) f_L(s)ds = \int_{\eta_L+\frac{\beta_L}{\xi_L}}^{K}(K-s)\frac{\theta_L}{\beta_L}(1 - \xi \frac{s-\eta_L}{\beta_L})^{-1-\frac{1}{\xi}}ds $

Let $x=\frac{s-\eta_L}{\beta_L}$, the equation above can be transformed to 

$\frac{p(T,K)}{d(T)} = \frac{\theta_L}{\beta_L} \int_{\frac{1}{\xi_L}}^{\frac{K-\eta_L}{\beta_L}}(K-\eta_L-\beta_L x)(1-\xi x)^{-1-\frac{1}{\xi}}dx$

$\frac{p(T,K)}{d(T)} =  \frac{\theta_L}{\beta_L}\left((K-\eta_L-\beta_L x)(1-\xi x)^\frac{-1}{\xi} -\beta_L(\frac{(1-\xi x)^{\frac{-1}{\xi}+1}}{\xi - 1})\right)\Big|_{\frac{1}{\xi_L}}^{\frac{K-\eta_L}{\beta_L}}$

Simplify

$\frac{p(T,K)}{d(T)} = \frac{\theta_L}{\beta_L} (1-\xi x)^{\frac{-1}{\xi}} \frac{(K-\eta)(\xi-1) + \beta_L(x-1)}{\xi-1}\Big|_{\frac{1}{\xi_L}}^{\frac{K-\eta_L}{\beta_L}}$

$=\frac{\theta_L}{\beta_L} (1-\xi \frac{K-\eta_L}{\beta_L})^{\frac{-1}{\xi}} \frac{(K-\eta_L)(\xi-1) + (K-\eta_L-\beta_L)}{\xi-1}$

Let $t=\theta, e = \eta, b=\beta, c= \xi$, validate by integration through Python


```python
from sympy import *
t, h, k, e, b, c, x = symbols('t h k e b c x')

pl = integrate((k-e-b*x)*(1-c*x)**(-1-1/c),x)
pl.simplify()
```


(2) $K \ge \eta_U$

$\frac{p(T,K)}{d(T)} - K + F(T) = \int_{K}^{\infty} (s-K) f_U(s)ds $

Let $x=\frac{s-\eta_U}{\beta_U}$, the equation above can be transformed to 

$\frac{p(T,K)}{d(T)} - K + F(T) = \int_{\frac{K-\eta_U}{\beta_U}}^{\infty}(s-K)\frac{\theta_U}{\beta_U}(1 + \xi \frac{s-\eta_U}{\beta_U})^{-1-\frac{1}{\xi}}ds = \frac{\theta_U}{\beta_U} \int_{\frac{K-\eta_U}{\beta_U}}^{\infty}(\eta_U - K + \beta_U x)(1 + \xi x)^{-1-\frac{1}{\xi}}dx$


$\frac{p(T,K)}{d(T)} - K + F(T) =  \frac{\theta_U}{\beta_U}\left(-(\eta_U-K+\beta_U x)(1+\xi x)^\frac{-1}{\xi} + \beta_U(\frac{(1+\xi x)^{\frac{-1}{\xi}+1}}{1- \frac{1}{\xi}})\right)\Big|_{\frac{K-\eta_U}{\beta_U}}^{\infty}$

$\frac{p(T,K)}{d(T)} - K + F(T) =  \frac{\theta_U}{\beta_U}(1+\xi x)^{\frac{-1}{\xi}}\frac{(K-\eta)(\xi-1) + \beta_U(1+x)}{\xi-1}\Big|_{\frac{K-\eta_U}{\beta_U}}^{\infty}$

$\frac{p(T,K)}{d(T)}=K - F(T)-\frac{\theta_U}{\beta_U}(1+\frac{\xi(K-\eta_U)}{\beta_U} )^{\frac{-1}{\xi}}\frac{(K-\eta)(\xi-1) + \beta_U + K - \eta_U }{\xi-1}$

Let $t=\theta, e=\eta, b=\beta, c= -1 - \frac{1}{\xi}$

```python
#load prices and filter
#Improve filter
import warnings
warnings.filterwarnings('ignore')
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d
from py_vollib.black_scholes.implied_volatility import implied_volatility as iv
from py_vollib.black_scholes import black_scholes
import numpy as np
import datetime

def load_implied():
    fileDt = 'HW2Data/date.csv'
    raw_dates = pd.read_csv(fileDt, sep='\t', lineterminator='\n')
    dfImplied = raw_dates.T
    dfImplied.columns = ['DiscountFactor','Foward','AdjClose'] 
    dfImplied = dfImplied.astype(float)
    dfImplied.index = pd.to_datetime(dfImplied.index)
    return dfImplied

dfImplied = load_implied()

def load_prices():
    filePrices = 'HW2Data/pricebystrike.csv'
    prices = pd.read_csv(filePrices, sep='\t', lineterminator='\n')
    dfPrices = prices.set_index('strike')
    dfPrices = dfPrices.astype(float)
    return dfPrices

dfPrices = load_prices()
impliedDiv = np.zeros(len(dfPrices.columns)) 
impliedDiv[5] = 0.9
impliedDiv[6] = 4.6
impliedDiv[7] = 0.9
impliedDiv[8] = 0.2
impliedDiv[9] = 0.2
impliedDiv
dfRate = -np.log(dfImplied.DiscountFactor) / ((datetime.datetime(2021, 9, 17) - dfImplied.index).days/365)

#for dt in dfPrices.columns:
def clean_up(dfPrices, dfImplied, impliedDiv):
    #init return dataframes    
    dfPricesCalib = pd.DataFrame(0, columns=dfPrices.columns, index=dfPrices.index)
    dfImpliedVol = pd.DataFrame(0, columns=dfPrices.columns, index=dfPrices.index)
    dfFilter = pd.DataFrame(False, columns=dfPrices.columns, index=dfPrices.index)
    
    ind = 0
    
    plt.figure(figsize=(16, 10))
    for dt in dfPrices.columns:
        price_date = dt
        dt_price_date = datetime.datetime.strptime(price_date, "%Y-%m-%d")
        dfIV = dfPrices[[price_date]]#[dfPrices.iloc[:,0]>0]
        dfIV = dfIV[dfIV.iloc[:,0]>0]
        dfIV.columns=[[price_date]]
        dfIV['IV'] = 0
        np_price = dfIV.iloc[:,0].to_numpy()
        S = dfImplied.AdjClose[dt_price_date] + impliedDiv[ind]
        flag = 'p'
        t = (datetime.datetime(2021, 9, 17) - dt_price_date).days/365
        r = dfRate[dt_price_date]
        for index, row in dfIV.iterrows():
            dfIV.loc[index,'IV'] = iv(row[0],S,index,t,r,flag)
        dfIV.columns = dfIV.columns.get_level_values(0)
        plt.plot(dfIV.index, dfIV.IV, label=dt)    
        fx = np.poly1d(np.polyfit(dfIV.index, dfIV.IV, 50))
        fxv = fx(dfIV.index)
        fpv = np.zeros(len(fxv))
        for ix in np.arange(len(fxv)):
            xiv = fxv[ix]
            index = dfIV.index[ix]
            fpv[ix] = black_scholes(flag,S,index,t,r,xiv)
        mu = np.mean(fxv - dfIV.IV)
        std = np.std(fxv - dfIV.IV)
        filter = np.abs(fxv - dfIV.IV) > 2*std
        #print("filtered {} points on date {}".format(sum(filter), dt))
        fit_label = "{} fit".format(dt)
        dfImpliedVol[dt] = pd.Series(data=fxv, index=dfIV.index)
        dfPricesCalib[dt] = pd.Series(data=fpv, index=dfIV.index)
        
        dfFilter[dt] = pd.Series(data=(np.abs(fxv - dfIV.IV) > 2*std), index=dfIV.index)
        plt.plot(dfIV.index, fxv, label=fit_label)
        ind = ind + 1
    #plt.title('Vol by trading date', fontsize=16)
    #plt.legend()
    #plt.grid()
    #plt.show()
    return [dfPricesCalib, dfImpliedVol, dfFilter]
[dfPricesCalib, dfImpliedVol, dfFilter] = clean_up(dfPrices, dfImplied, impliedDiv)
```

```python
eta_lower = 11000
eta_upper = 17000
```

```python
import scipy
from scipy.optimize import minimize
from numpy.random import rand

#theta = [s, r, K, sigma, T] #spot price, rate, strike, sigma, Time to expire

dfPutInputs = dfPricesCalib[['2021-09-02']].copy()
dfPutInputs = dfPutInputs[dfPutInputs.index <= eta_lower]
dfPutInputs = dfPutInputs.dropna()
dfPutInputs.columns=['2021-09-02']
discFactor = dfImplied.DiscountFactor['2021-09-02']
#dfPutInputs

#add discount
def p_lower_fx(theta, beta, xi, K) : 
    price = discFactor * (theta/beta*(1-xi*(K-eta_lower)/beta)**(-1/xi))*((K-eta_lower)*(xi-1)+(K-eta_lower-beta))/(xi-1)
    return price
#*((k-eta_lower)(c-1)+(K-eta_lower-beta))/(c-1) 

x = [0.2,1500,-0.2]

# objective function
def objective_L(x):
    [theta, beta, xi] = x
    opt_size = len(dfPutInputs)
    sum_mse = np.zeros(opt_size)
    price_date = dfPutInputs.columns[0]
    dt_price_date = datetime.datetime.strptime(price_date, "%Y-%m-%d")
    #t = (datetime.datetime(2021, 9, 17) - dt_price_date).days/365
    #r = dfRate[dt_price_date]
    #print(t,r)
    for ix in np.arange(opt_size):
        #px_calc = black_scholes('p',15604.25,dfPutInputs.index[ix],t,r,sigma)
        px_calc = p_lower_fx(theta,beta,xi,dfPutInputs.index[ix])
        sum_mse[ix] = np.square(px_calc - dfPutInputs.iloc[ix])
        #print( px_calc, dfPutInputs.iloc[ix])

    return np.sum(sum_mse)/opt_size
# define range for input
#r_min, r_max = 0.1, 1.0
# define the starting point as a random sample from the domain
#pt = r_min + rand(1) * (r_max - r_min)
#print("Initial params are : {}".format(pt[0]))
#objective_L(x_list)
# perform the search
result = minimize(objective_L, x, method='nelder-mead', bounds= scipy.optimize.Bounds([0, 1, -1], [1.0, 5000, 0]))
print('Status : %s' % result['message'])
print('Total Evaluations: %d' % result['nfev'])
# evaluate solution
solution = result['x']
evaluation = objective_L(solution)
print('Solution: f(%s) = %.4f' % (solution, evaluation))
```

```python
dfPutInputsU = dfPricesCalib[['2021-09-02']].copy()
dfPutInputsU = dfPutInputsU[dfPutInputsU.index >= eta_upper]
dfPutInputsU = dfPutInputsU.dropna()
dfPutInputsU.columns=['2021-09-02']
Fwd = dfImplied.Foward['2021-09-02']
dfPutInputsU

def p_upper_fx(theta, beta, xi, K) : 
    price = discFactor * (K-Fwd - (theta/beta*(1+xi*(K-eta_upper)/beta)**(-1/xi))*((K-eta_upper)*(xi-1)+(K-eta_upper+beta))/(xi-1))
    return price
#*((k-eta_lower)(c-1)+(K-eta_lower-beta))/(c-1) 

xu = [0.01,12350,0.9]

# objective function
def objective_U(x):
    [theta, beta, xi] = x
    opt_size = len(dfPutInputsU)
    sum_mse = np.zeros(opt_size)
    price_date = dfPutInputsU.columns[0]
    dt_price_date = datetime.datetime.strptime(price_date, "%Y-%m-%d")
    #t = (datetime.datetime(2021, 9, 17) - dt_price_date).days/365
    #r = dfRate[dt_price_date]
    #print(t,r)
    for ix in np.arange(opt_size):
        #px_calc = black_scholes('p',15604.25,dfPutInputs.index[ix],t,r,sigma)
        px_calc = p_upper_fx(theta,beta,xi,dfPutInputsU.index[ix])
        sum_mse[ix] = np.square(px_calc - dfPutInputsU.iloc[ix])
        #print( px_calc, dfPutInputs.iloc[ix])

    return np.sum(sum_mse)/opt_size

# perform the search
result = minimize(objective_U, xu, method='nelder-mead', bounds= scipy.optimize.Bounds([0, 1, 0], [1.0, 15000, 1]))
print('Status : %s' % result['message'])
print('Total Evaluations: %d' % result['nfev'])
# evaluate solution
solution = result['x']
evaluation = objective_U(solution)
print('Solution: f(%s) = %.4f' % (solution, evaluation))
```
