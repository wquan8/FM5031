```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
x = np.linspace(-5, 5, 100)
#Generating example data

y = x**3 - 1 + 3*np.random.normal(size = 100)

z = np.polyfit(x, y, 3)

fx = np.poly1d(z)
t = fx(x)
print(x.shape)
print(t)
print(t.shape)
plt.plot(x,y, label='orig')
plt.plot(x, t, label = 'poly')
plt.legend()
plt.show()
```
